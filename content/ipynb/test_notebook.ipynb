```python
import inspect

class Sketchy(object):

    @property
    def word(self):
        caller = inspect.stack()[1][3]
        if "dodgy" in caller:
            return '"{0}"!'.format(caller)
        else:
            return self._word

    @word.setter
    def word(self, val):
        self._word = val
```

```python
sketchy = Sketchy()
sketchy.word = "Hello,"
```
